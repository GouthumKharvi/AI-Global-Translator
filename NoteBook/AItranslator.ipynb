# Loading the Input

```python
import sounddevice as sd
from scipy.io.wavfile import write

def record_audio(duration=15, filename=r"C:\Users\Gouthum\Downloads\AI Translator\scripts\input.wav"):
    print(f"Recording for {duration} seconds... Speak now!")
    fs = 16000  # Sample rate
    recording = sd.rec(int(duration * fs), samplerate=fs, channels=1)
    sd.wait()  # Wait until recording is finished
    write(filename, fs, recording)
    print("Recording saved as", filename)

record_audio()
```

# VOICE TO TEXT (TRANSALATOR)

```python
import whisper
import argostranslate.package
import argostranslate.translate

# Load Whisper model
model = whisper.load_model("base")  # or "small" / "medium" / "large"

# Transcribe audio
result = model.transcribe(r"C:\Users\Gouthum\Downloads\AI Translator\scripts\input.wav")
english_text = result["text"]
print("Transcribed English Text:\n", english_text)

# Setup Argos Translate for English to Hindi (example)
from_code = "en"
to_code = "fr"  # Use "kn" for Kannada, "ta" for Tamil, etc.

# Load available translation packages
installed_languages = argostranslate.translate.get_installed_languages()
from_lang = next(filter(lambda x: x.code == from_code, installed_languages))
to_lang = next(filter(lambda x: x.code == to_code, installed_languages))

# Translate
translated_text = from_lang.get_translation(to_lang).translate(english_text)
print(f"\nTranslated Text ({from_code} âžœ {to_code}):\n", translated_text)
```

```python
import whisper
from deep_translator import GoogleTranslator

# Step 1: Transcribe audio using Whisper
model = whisper.load_model("base")
result = model.transcribe(r"C:\Users\Gouthum\Downloads\AI Translator\scripts\input.wav")
english_text = result["text"]
print("Transcribed English Text:\n", english_text)

# Step 2: Translate to desired language (e.g., Tamil = 'ta')
translated_text = GoogleTranslator(source='auto', target='ta').translate(english_text)
print("\nTranslated Text (to Tamil):\n", translated_text)
```
